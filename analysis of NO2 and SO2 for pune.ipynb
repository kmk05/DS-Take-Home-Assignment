```python
import matplotlib.pyplot as plt
```

```python
NO2=[45,50,68,87,77,83,101,62]
SO2=[19,23,22,25,20,35,46,16]
value2=["2013","2014","2015","2016","2017","2018","2019","2020"]
```

```python
plt.scatter(value2,NO2)
plt.scatter(value2,SO2)
plt.plot(value2, NO2,label='NO2')
plt.plot(value2,SO2,label='SO2')
plt.legend()
plt.savefig('plot.png')
plt.title("NO2 AND SO2 Values")
plt.xlabel("Year")
plt.ylabel("Concentration in µg/m3")
plt.show()
```
